### Preguntas e hipótesis de interés

Para guiar el análisis eficazmente, se formularon las siguientes **preguntas e hipótesis**:

1. **¿Qué características del cliente predicen mejor la baja del servicio de tarjeta de crédito?**

*Hipótesis:* Los clientes con límites de crédito más bajos y con menores transacciones son más propensos a darse de baja debido a una menor interacción y satisfacción.


2. **¿Influye la edad o la antigüedad del cliente en el banco en la decisión de darse de baja?**

*Hipótesis:* Los clientes más jóvenes y con menor antigüedad son más propensos a darse de baja, posiblemente, debido a la falta de compromiso a largo plazo o menor lealtad.

3. **¿Los clientes con mayor uso de tarjetas y saldos rotativos tienden a quedarse?**
*Hipótesis:* Los clientes con mayor uso de tarjetas tienen menores probabilidades de abandonar la cuenta de la tarjeta de crédito, ya que dependen mucho de sus gastos.

4. **¿Existe una relación entre la pérdida de clientes y el contacto con el servicio de atención al cliente?**
*Hipótesis:* Los clientes que contactan al servicio de atención al cliente con mayor frecuencia tienen mayor probabilidad de dar la baja de la tarjeta de crédito debido a problemas no resueltos o insatisfacción.

5. **¿El nivel de ingresos afecta la probabilidad de abandono?**
*Hipótesis:* Los clientes con ingresos más bajos pueden ser más sensibles a las comisiones o a los cambios en las condiciones de la cuenta de tarjeta de crédito, lo que resulta en mayores tasas de abandono.

Estas hipótesis se explorarán mediante resúmenes estadísticos y visualizaciones en la fase de análisis de datos extraídos (EDA) para identificar los principales factores de abandono.

```python
# 3. Violin Plot: Promedio Tasa de utilización por categoría de ingresos y abandono
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Income_Category', y='Avg_Utilization_Ratio', hue='Attrition_Flag',
               palette='pastel', split=True)
plt.title("Tasa de utilización por categoría de ingresos y abandono")
plt.xlabel("Categoría de Ingresos")
plt.ylabel("Promedio tasa de utilización")
plt.legend(title='Attrition Flag', labels=['Existentes', 'Desertores'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Insights:

```python
# 4. Mapa de calor categórico: Tasa de abandono por Nivel educativo y categoría de tarjeta
# Step 1: Calcular cata de abandono en tabla pivot
pivot = pd.crosstab(df['Education_Level'], df['Card_Category'], values=df['Attrition_Flag'], aggfunc='mean')

# Step 2: Plot del Mapa de calor
plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("Tasa de abandono por Nivel educativo y categoría de tarjeta")
plt.xlabel("Categoría Tarjeta")
plt.ylabel("Nivel Educativo")
plt.tight_layout()
plt.show()
```

Insights:

La tasa de abandono es baja en todas las categorías, pero:

- Los titulares de tarjetas "Platinum" que poseen educación superior (doctorado, posgrado) muestran una tasa de abandono mucho mayor.

```python
# Mostramos la variable objetivo codificada
df['Attrition_Flag'].value_counts()
```

Como lo habíamos realizado en la sección "Codificar la variable objetivo"; mostramos, nuevamente, el resultado de dicha codificación.

```python
# Mostrar one-hot encoding variables categóricas
df = pd.get_dummies(df, columns=['Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category'], drop_first=True)
```

Aplicamos One-Hot Encoding a través de la función get_dummies de manera que podamos preparar los datos categóricos para los modelos de Machine Learning a emplear puesto que requieren de entradas numéricas.

```python
#Creaación de nueva feature: EngagementScore
df['EngagementScore'] = df['Total_Trans_Ct'] * df['Total_Ct_Chng_Q4_Q1']
```

Se agrega la nueva feature "EngagementScore" la cual combina dos métricas interesantes vinculadas a la actividad transaccional en un único valor con el objetivo de medir la interacción del cliente a largo del tiempo.

## Feature Selection

```python
# Definir variable objetivo
target = 'Attrition_Flag'

# Definir features (excluir objetivo)
features = [col for col in df.columns if col != target]

# Separación de features y objetivos
X = df[features]
y = df[target]

# Muestreo del shape y primeras filas
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
print("\nFirst few rows of X:")
print(X.head())
print("\nFirst few rows of y:")
print(y.head())
```

Se define la variable objetivo y se genera la lista "features" con los nombres de columna que NO sean igual a la variable objetivo. Creamos los dataframes "X" e "Y" y mostramos los primeros resultados.
Al tener definidos el objetivo y feature, el próximo paso será la gestión del desbalance de clases y el entrenamiento de los modelos de clasificación.



## Preprocesamiento de datos

```python
#Preprocesamiento de Datos - Separación, estandarización, Standardize y gestión del desbalance
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Identificación de columnas numéricas para la estandarización
numeric_cols = [
    'Customer_Age', 'Dependent_count', 'Months_on_book', 'Total_Relationship_Count',
    'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Credit_Limit',
    'Total_Revolving_Bal', 'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1',
    'Total_Trans_Amt', 'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1',
    'Avg_Utilization_Ratio', 'EngagementScore'
]

scaler = StandardScaler()
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])
```

```python
# Aplicar SMOTE
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

# Salida del shape y la clase balanciada
print("Shape of X_train_resampled:", X_train_balanced.shape)
print("Shape of y_train_resampled:", y_train_balanced.shape)
```

## Modelos

### Modelo 1: Logistic Regression (con Cross-Validation)

Este modelo usa class_weight = 'balanced' para la gestión del desbalance. A su vez, aplica 5-fold-validation para su evaluación.
Las métricas que devuelve son:
- Accuracy
- Precision
- Recall
- F1
- ROC AUC

#### Librerías necesarias para la implementación

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

#### División de datos en conjuntos de entrenamiento y prueba

```python
# Inicialización de Logistic Regression con balanced class weights
logreg = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)

# Ajustar a datos balanceados con SMOTE
logreg.fit(X_train_balanced, y_train_balanced)
```

#### Predicción en conjunto de prueba

```python
# Predicción en el set de test
y_pred = logreg.predict(X_test)
y_proba = logreg.predict_proba(X_test)[:, 1]
```

#### Evaluación del rendimiento del modelo

```python
# Métricas de evaluación
print("🔹 Reporte de Logistic Regression Classification:")
print(classification_report(y_test, y_pred))

print("🔹 Matriz de confusión:")
print(confusion_matrix(y_test, y_pred))

print("🔹 ROC AUC Score:", roc_auc_score(y_test, y_proba))

# Cross-validation ROC AUC
cv_auc = cross_val_score(logreg, X_train_balanced, y_train_balanced, cv=5, scoring='roc_auc')
print("🔹 Cross-Validation ROC AUC:", cv_auc.mean())
```

Este resultado muestra qué tan bien Logistic Regression gestiona la tarea de la clasificación de abandono, si sobreestima o subestima la tasa de abandono y si es adecuado como referencia para compararse con modelos más complejos.

### Modelo 2: Random Forest

Este modelo gestiona bien tanto los datos númericos como los categóricos. Además, resiste el "overfitting" debido a la naturaleza del conjunto y proporciona la importancia de los features para poder interpretar a los impulsores de abandono.

#### Librerías necesarias para la implementación

```python
from sklearn.ensemble import RandomForestClassifier
```

#### División de datos en conjuntos de entrenamiento y prueba

```python
# Inicialización del modelo con class_weight='balanced' para mejorar el desbalanceo
rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)

# Ajustar en datos de entrenamiento balanceados con SMOTE
rf.fit(X_train_balanced, y_train_balanced)
```

#### Predicción en conjunto de prueba

```python
# Predicción en el set de testeo
y_pred_rf = rf.predict(X_test)
y_proba_rf = rf.predict_proba(X_test)[:, 1]
```

#### Evaluación del rendimiento del modelo

```python
# Métricas de evaluación
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.model_selection import cross_val_score

print("🔹 Reporte de Random Forest Classification:")
print(classification_report(y_test, y_pred_rf))

print("🔹 Matriz de confusión:")
print(confusion_matrix(y_test, y_pred_rf))

print("\n ROC AUC Score:", roc_auc_score(y_test, y_proba_rf))

cv_auc_rf = cross_val_score(rf, X_train_balanced, y_train_balanced, cv=5, scoring='roc_auc')
print("\n Cross-validated ROC AUC:", cv_auc_rf.mean())
```

Insights: ¿Random Forest supera a Logistic Regression en cuanto a recuperación o AUC? La importancia de las features puede ayudar a explicar los factores que conducen a la pérdida de clientes.

### Modelo 3: XGBoost

Este modelo tiene excelente rendimiento en datasets tabulares. Además, maneja correctamente los valores faltantes, la no linealidad y las interacciones. Puede usar la función "scale_pos_weight" para corregir el desequilibrio de clases.

#### Librerías necesarias para la implementación

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.model_selection import cross_val_score
```

#### División de datos en conjuntos de entrenamiento y prueba

```python
# Inicializar XGBoost con el desequilibrio de clases gestionado mediante scale_pos_weight
# Estimación: proporción de clases mayoritarias/minoritarias (~8500 / 1600 ≈ 5,3)
xgb = XGBClassifier(scale_pos_weight=5.3, use_label_encoder=False, eval_metric='logloss', random_state=42)

# Ajustar en datos de entrenamiento balanceados con SMOTE
xgb.fit(X_train_balanced, y_train_balanced)
```

#### Predicción en conjunto de prueba

```python
# Predicción en el set de testeo
y_pred_xgb = xgb.predict(X_test)
y_proba_xgb = xgb.predict_proba(X_test)[:, 1]
```

#### Evaluación del rendimiento del modelo

```python
# Métricas de evaluación
print("\n Reporte XGBoost Classification:")
print(classification_report(y_test, y_pred_xgb))

print("\n Matriz de confusión:")
print(confusion_matrix(y_test, y_pred_xgb))

print("\n ROC AUC Score:", roc_auc_score(y_test, y_proba_xgb))

# Cross-validation
cv_auc_xgb = cross_val_score(xgb, X_train_balanced, y_train_balanced, cv=5, scoring='roc_auc')
print("\n Cross-validated ROC AUC:", cv_auc_xgb.mean())
```

¿Qué es lo que se aprenderá?:

- Un candidato firme para el modelo de mejor rendimiento antes del ajuste de hiperparámetros.

## Optimización de los Modelos

Se aplicará "GridSearchCV" o "RandomizedSearchCV" con "Cross-validation" en XGBoost.
GridSearchCV prueba múltiples combinaciones de parámetros, utiliza cross-validation para evaluar cada configuración y ayuda a maximizar el rendimiento de la generalización con datos no analizados.

#### Parámetros para XGBoost

```python
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.3],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
}
```

#### Importaciones y aplicaciones para XGBoost y GridSearchCV

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

# Modelo base con control para el desbalanceo
xgb_model = XGBClassifier(
    scale_pos_weight=5.3,
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

# Definir parámetros del grid
param_grid = {
    'n_estimators': [100,200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.3],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
}

# Setup GridSearchCV
grid_search = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    scoring='roc_auc',
    n_jobs=-1,
    cv=5,
    verbose=1
)

# Correr grid search en los datos de entrenamiento con SMOTE balanceado
grid_search.fit(X_train_balanced, y_train_balanced)

# Salida de los mejores parámetros y scores
print("Mejores parámetros encontrados:", grid_search.best_params_)
print("Mejor validación cruzada ROC AUC:", grid_search.best_score_)
```
