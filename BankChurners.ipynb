### Preguntas e hipótesis de interés

Para guiar el análisis eficazmente, se formularon las siguientes preguntas e hipótesis:

*   ¿Qué artistas muestran un dominio constante en las cifras de streaming? Un pequeño grupo de artistas logra sistemáticamente mayores cifras de streaming, lo que sugiere fuertes efectos de fidelización de la marca o de la audiencia.

*   ¿Existen patrones estacionales notables que influyan en el éxito del streaming de canciones? Las canciones lanzadas durante períodos específicos, como los meses de verano, obtienen mayores volúmenes de streaming en comparación con otras temporadas.

*   ¿Las canciones colaborativas (con varios artistas) alcanzan un mayor número de reproducciones en comparación con las canciones solistas? Hipótesis: Las canciones colaborativas atraen a una audiencia más amplia debido a la combinación de seguidores, lo que resulta en un número de reproducciones significativamente mayor que las canciones solistas.

*   ¿El contenido explícito está asociado con el aumento o la disminución de la popularidad de las reproducciones en Spotify en 2024? Hipótesis: Las canciones etiquetadas como contenido explícito generalmente reciben un mayor número de reproducciones, posiblemente debido a su atractivo para un público más joven o a segmentos demográficos específicos del mercado.

#### Popularidad en Spotify en relación a mes de lanzamiento

```python
# Convert to datetime if not already
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')

# Extract month name
df['Release Month'] = df['Release Date'].dt.month_name()

month_translation = {
    'January': 'Enero', 'February': 'Febrero', 'March': 'Marzo',
    'April': 'Abril', 'May': 'Mayo', 'June': 'Junio',
    'July': 'Julio', 'August': 'Agosto', 'September': 'Septiembre',
    'October': 'Octubre', 'November': 'Noviembre', 'December': 'Diciembre'
}

df['Mes de Lanzamiento'] = df['Release Month'].map(month_translation)

meses_ordenados = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
                   'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

df['Mes de Lanzamiento'] = pd.Categorical(df['Mes de Lanzamiento'], categories=meses_ordenados, ordered=True)

plt.figure(figsize=(12, 6))
sns.boxplot(x='Mes de Lanzamiento', y='Spotify Popularity', data=df, palette='Set3')
plt.title('Popularidad en Spotify por Mes de Lanzamiento')
plt.xticks(rotation=45)
plt.xlabel('Mes de Lanzamiento')
plt.ylabel('Popularidad en Spotify')
plt.grid(True)
plt.show()
```

Insight: A primera vista, se observa que abril, mayo y marzo son los meses en donde la mediana o los cuartiles superiores son más elevados. Particularmente, enero tiene una mediana elevada pero no así su cuartil superior. Esto apoya la teoría que existen patrones estacionales notables que hacen que el lanzamiento de una canción sea más propenso en determinados meses del año. Particularmente, se ve que coincide con la primavera del hemisferio norte y el otoño del hemisferio sur.

### Multivariados

#### Análisis de lanzamiento de canciones con contenido explícito

```python
# Conversión a datetime si no está listo
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')

# Extracción del nombre del mes
df['Release Month'] = df['Release Date'].dt.month_name()

month_translation = {
    'January': 'Enero', 'February': 'Febrero', 'March': 'Marzo',
    'April': 'Abril', 'May': 'Mayo', 'June': 'Junio',
    'July': 'Julio', 'August': 'Agosto', 'September': 'Septiembre',
    'October': 'Octubre', 'November': 'Noviembre', 'December': 'Diciembre'
}

df['Mes de Lanzamiento'] = df['Release Month'].map(month_translation)

meses_ordenados = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
                   'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

df['Mes de Lanzamiento'] = pd.Categorical(df['Mes de Lanzamiento'], categories=meses_ordenados, ordered=True)

# Asegurar que ambas columnas existan con el nombre correcto
explicit_map = {0: 'No', 1: 'Sí'}
df['Explicit Track'] = df['Explicit Track'].map(explicit_map)

plt.figure(figsize=(12, 6))
sns.boxplot(x='Mes de Lanzamiento', y='Spotify Popularity', hue='Explicit Track', data=df, palette='pastel')
plt.title('Popularidad en Spotify por Mes de Lanzamiento')
plt.xticks(rotation=45)
plt.xlabel('Mes de Lanzamiento')
plt.ylabel('Popularidad en Spotify')
plt.grid(True)
plt.show()
```

Insights: Vemos que agosto, noviembre y diciembre son meses en donde el lanzamiento de las canciones explícitas superan a las que no lo tienen. Es información interesante a tener en cuenta para un artista que compone canciones con contenido explícito y desea saber cuáles son los meses con mejores recepciones para el lanzamiento.
